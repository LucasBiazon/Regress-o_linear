```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib as plt

X = np.array([[1, 1, 20, 2, 1, 1],
              [1, 1, 40, 2, 1, 1],
              [1, 2, 30, 2, 1, 2],
              [1, 2, 20, 2, 2, 2],
              [1, 3, 30, 2, 2, 3],
              [1, 3, 40, 3, 2, 3],
              [1, 4, 20, 3, 3, 1],
              [1, 4, 30, 3, 3, 1],
              [1, 5, 40, 3, 3, 2],
              [1, 5, 30, 3, 3, 2],
              [1, 6, 30, 3, 3, 3],
              [1, 6, 30, 3, 4, 3],
              [1, 6, 30, 3, 2, 4],
              [1, 3, 30, 3, 3, 4]])

Y = np.array([[7.9],[ 6.6], [6.4], [8.6], [6.8], [6.8], [9.1], [8.2], [6.1], [7.3], [5.8], [8.1], [3.4], [8.9]])

X_one = X
Xt = np.transpose(X)         
XtX = np.matmul(Xt,X)         
XtX_inv = np.linalg.inv(XtX)  
XtY = np.matmul(Xt,Y)         
coef = np.matmul(XtX_inv,XtY) 
Yprev = np.matmul(X_one,coef)

plt.xlabel("Celular hr")
plt.ylabel("Hora de sono")
plt.title("Hora de sono x Celular hr")
plt.scatter(X[:,1],Y,c = 'blue', s=15, alpha=0.5)
plt.show()

print("")

plt.xlabel("Açúcar")
plt.ylabel("Hora de sono")
plt.title("Hora de sono x Açucar")
plt.scatter(X[:,2],Y,c = 'blue', s=15, alpha=0.5)
plt.show()

print("")

plt.xlabel("Agua")
plt.ylabel("Hora de sono")
plt.title("Hora de sono x agua")
plt.scatter(X[:,3],Y,c = 'blue', s=15, alpha=0.5)
plt.show()

print("")

plt.xlabel("Atividade física")
plt.ylabel("Hora de sono")
plt.title("Hora de sono x Atividade física")
plt.scatter(X[:,4],Y,c = 'blue', s=15, alpha=0.5)
plt.show()

print("")

plt.xlabel("Café")
plt.ylabel("Hora de sono")
plt.title("Hora de sono x Café")
plt.scatter(X[:,5],Y,c = 'blue', s=15, alpha=0.5)
plt.show()

print("")
plt.scatter(Y,Yprev,c = 'blue', s=15, alpha=0.5)
plt.xlabel('Prob real')
plt.ylabel('Prob prev')
plt.show()

df = pd.DataFrame(np.hstack((X, Y)), columns=['Intercepto', 'Celular', 'Açúcar', 'Água', 'Atividade', 'Café', 'Horas_de_Sono'])

print(df)

correlation_matrix = df.corr()


sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', square=True)
X_novo = np.array([[1, 1, 40, 3, 4, 3]])  


previsao_horas = np.matmul(X_novo, coef)
print(f"Previsão das horas de sono: {previsao_horas[0][0]:.2f} horas")
```
